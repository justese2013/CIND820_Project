```python
# set categorical data
df['Geography'] = df['Geography'].astype('category')
df['Immig.'] = df['Immig.'].astype('category')
df['Dur. of unempl'] = df['Dur. of unempl'].astype('category')
df['Age group'] = df['Age group'].astype('category')
```

```python
# Strip White spaces in all categorical columns
df['Geography'] = df['Geography'].str.strip()
df['Immig.'] = df['Immig.'].str.strip()
df['Dur. of unempl'] = df['Dur. of unempl'].str.strip()
df['Age group'] = df['Age group'].str.strip()
```

```python
df['Dur. of unempl'].value_counts()
```

```python
#from numpy.ma.core import ceil

#set float values to int
df['Both sexes'] = df['Both sexes'].astype(int)
df[' Males'] = df[' Males'].astype(int)
df[' Females'] = df[' Females'].astype(int)

print(df)
```

```python
# Create a new column for 'Dur of employment' summarizing its attributes
Dur_of_unempl = []
for row in df['Dur. of unempl']:
        if row == '1 - 4 weeks' :  Dur_of_unempl.append('Short duration')
        elif row == '5 - 13 weeks':   Dur_of_unempl.append('Short duration')
        elif row == '14 - 25 weeks':   Dur_of_unempl.append('Medium duration')
        elif row == '26 weeks ':   Dur_of_unempl.append('Medium duration')
        elif row == '27 weeks or more':   Dur_of_unempl.append('Long duration')
        elif row == '27 - 51 weeks':   Dur_of_unempl.append('Long duration')
        elif row == '52 weeks':   Dur_of_unempl.append('Long duration')
        elif row == '53 weeks or more':   Dur_of_unempl.append('Long duration')
        elif row == 'Duration unknown':   Dur_of_unempl.append('Duration Unknown')
        elif row == 'Average weeks unemployed (no top-code)':  Dur_of_unempl.append('Long duration')
        elif row == 'Average weeks unemployed (top-code = 99 weeks)':  Dur_of_unempl.append('Long duration')
        
        
        
        else:           Dur_of_unempl.append('Duration Unknown')

```

```python
df['Dur_of_unempl'] = Dur_of_unempl
print(df)
df['Dur_of_unempl'].value_counts()
```

```python
# We shall begin our exploratory data analysis to check for features that will help us predict the duration of unemploymet
```

```python
# Since the categorical variables in the data set have not be coded with numerical values we will be using the ONE-HOT (NUMERICAL) ENCODING
# We will first create a list of all the categotical variable
cat_cols=['Immig.','Age group','Geography', 'Dur. of unempl']
# Next Create a copy of the data frame
df_onehot= df
df_onehot
# Next convert only categorical variables to dummy/one-hot features
df_onehot = pd.get_dummies(df, columns=cat_cols, prefix = cat_cols)
#print the dataset
df_onehot
```

```python
# Now split data set into train/test set split
from sklearn.model_selection import train_test_split
class_col_name='Dur_of_unempl'
one_hot_feature_names=df_onehot.columns[df_onehot.columns != class_col_name]
# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(df_onehot.loc[:, one_hot_feature_names], df_onehot[class_col_name], test_size=0.3,random_state=109) # 70% training and 30% test
```

```python
# Repeat Naive Bayes modeling
from sklearn.naive_bayes import MultinomialNB
```

```python
#Create a MultiNomial NB Classifier
nb= MultinomialNB()

#Train the model using the training sets
nb.fit(X_train, y_train)

#Predict the response for test dataset
y_pred = nb.predict(X_test)
print ("Succesfully done..")
```

```python
print("Number of features used ",nb.n_features_)
print("Classes ",nb.classes_)
print("Number of records for classes ",nb.class_count_)
print("Log prior probability for classes ", nb.class_log_prior_)
print("Log conditional probability for each feature given a class\n",nb.feature_log_prob_)
```

```python
from sklearn.metrics import confusion_matrix
cf=confusion_matrix(y_test, y_pred)
print ("Confusion Matrix")
print(cf)
tn, fp, fn, tp=cf.ravel()
print ("TP: ", tp,", FP: ", fp,", TN: ", tn,", FN:", fn)
```

```python
from sklearn.metrics import classification_report
from sklearn import metrics

print(classification_report(y_test, y_pred))
```

```python
#Decision tree classification
from sklearn import tree
clf = tree.DecisionTreeClassifier(max_depth=5)
clf = clf.fit(X_train, y_train)
import graphviz
#Get unique class values to display on the tree
class_values=df_onehot[class_col_name].unique()
print ("class Names",class_values)

dot_data = tree.export_graphviz(clf, out_file=None, 
                                feature_names=one_hot_feature_names,  
                                class_names=class_values,
                                filled=True)

# Draw graph
graph = graphviz.Source(dot_data, format="png") 
graph
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import classification_report
from sklearn import metrics

print(classification_report(y_test, y_pred))
```
